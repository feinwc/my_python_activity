```python
import pandas as pd

ott = pd.read_csv('ott_merge.csv')
mov = pd.read_excel('movies_merge.xlsx')

ott.shape
mov.shape
ott.head()
mov.head()
```

```python
# How many films from each year (released from 2012 to the present) were watched on Netflix?

total = pd.merge(ott, mov, on='ID', how='left')
total
```

```python
total_12 = total[total['Year'] > 2011]
total_12
```

```python
total_12.groupby('Year')[['Netflix']].sum()
```

```python
# Solution: Determine the number of movies watched on Netflix since 2012.
q1 = total.groupby(['Year'])['Netflix'].agg('sum').reset_index()
q1[q1['Year'] >= 2012]
```

```python
# Question 2: What is the average runtime of movies released each year?
q2 = total.groupby(['Year'])['Runtime'].agg('mean').reset_index()
q2[q2['Year'] >= 2012]
```

```python
# Question 3: What are the best and worst reviews that movies released each year received on Rotten Tomatoes?
q3 = total.groupby(['Year'])['Rotten Tomatoes'].agg(['max','min']).reset_index()
q3[q3['Year'] >= 2012]
```
