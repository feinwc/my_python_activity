```python
import pandas as pd

ott = pd.read_csv('ott_merge.csv')
mov = pd.read_excel('movies_merge.xlsx')

ott.shape
mov.shape
ott.head()
mov.head()
```

```python
# Merge
ott_mov = pd.merge(ott, mov, on='ID', how='left')
ott_mov
```

```python
# Q1 the film release date and content rating.

ott_mov.pivot(index='Title', columns='Age', values='Year')
```

```python
# Q2 the title of movies, the directors, and genres by content rating.
ott_mov.pivot(index='Title', columns='Age', values=['Directors', 'Genres'])
```

```python
# Q3 the title of movies, the released year, and language by content rating.
ott_mov.pivot(index='Title', columns='Age', values=['Year', 'Language'])
```

```python
# Q4 Netflix screened movies based on language, runtime, and country.
ott_mov.pivot(index='Title', columns='Netflix', values=['Language','Runtime','Country'])
```

```python
# Q5 the title of movies, specified language, potential runtime, and genres by content rating.
ott_mov.pivot(index='Title', columns='Age', values=['Language','Runtime','Genres'])
```
