```python
import pandas as pd
import numpy as np

ott = pd.read_csv('ott_merge.csv')
mov = pd.read_excel('movies_merge.xlsx')

ott.shape
mov.shape
ott.head()
mov.head()
```

```python
# Merge
ott_mov = pd.merge(ott, mov, on='ID', how='left')
ott_mov

# Subsetting: mov_ott_runtime consisting of the columns ID, Runtime and Genres. 
mov_ott_runtime = ott_mov[['ID','Runtime','Genres']]
mov_ott_runtime
```

```python
# What is the effect of adding 60 seconds (1 minute) to each movie?
# add 1 min by using add()
mov_ott_runtime['Runtime'].add(1)
```

```python
#Question 2: Which movies are documentaries?
#Create a column Gen_doc
mov_ott_runtime['Gen_doc'] = np.where(mov_ott_runtime['Genres'].str.contains('Documentary'),
                                      'Documentary','Not Documentary')

mov_ott_runtime
```

```python
# Use the applymap (determine length of string).
mov_ott_runtime.Gen_doc.apply(len)
```

```python
# suggests subtracting 6 seconds (0.01 minutes) from the Runtime of all the movies
mov_ott_runtime['Runtime'].subtract(0.01)
mov_ott_runtime
```

### lambda functions

```python
# Add 60 seconds or one minute with lambda function.
ott_mov['Runtime'] = ott_mov['Runtime'].apply(lambda x: x + 1)

# View output
ott_mov['Runtime']
```

```python
ott_mov['Runtime'] = ott_mov['Runtime'].apply(lambda x: x - 0.01)
ott_mov['Runtime']
```

```python
# Create a new DataFrame by submitting the mov_ott DataFrame and name it mov_ott_ratings
mov_ott_ratings = ott_mov[['ID','IMDb','Rotten Tomatoes']]
mov_ott_ratings
```

```python
# Replace the NaN values with 0 in both the IMDb and Rotten Tomatoes columns by using fillna() function
# Replace missing values with 0.
mov_ott_ratings_final = mov_ott_ratings.fillna(0)
mov_ott_ratings_final
```

```python
# 1. What is the average rating per movie?

def avg_rating(df1,df2):
    df = (df1/10 + df2)/2
    return df;

# Create a new column Rating and apply the user-defined function

mov_ott_ratings_final['Rating'] = avg_rating(mov_ott_ratings_final['IMDb'],mov_ott_ratings_final['Rotten Tomatoes'])
mov_ott_ratings_final
```

```python
# 2.How many movies were released per content rating (age)? (by using count() function)

# Categorical count.
def cnt(df1):
    print(df1.value_counts())
    
# Create a Series and Ensure the data type is category by specifying astype('category').)
df = ott_mov['Age'].astype('category')

# Number of movies released per 'Age'
cnt(df)
```

```python
# 3. How many movies were released per year?

def cnt(df1):
    print(df1.value_counts())
    
df = ott_mov['Year'].astype('category')
cnt(df)
```
