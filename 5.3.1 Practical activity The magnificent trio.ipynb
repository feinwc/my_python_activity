```python
# Import the necessary libraries.
import pandas as pd

# Import the Excel file.
transport = pd.read_excel('transport_costs.xlsx')

# View the DataFrame.
print(transport.shape)
print(transport.dtypes)
transport
```

```python
# Drop columns currency, distance unit.

transport_sub = transport[['region', 'country', 'port',
                            'sea freight cost', 'road transport cost per km']]

# View the output.
transport_sub.head()
```

```python
transport_sub['total_sea_freight_costs'] = transport_sub['sea freight cost'] * 2 * 4
transport_sub['total_road_transport_costs'] = transport_sub['road transport cost per km'] * 10000
transport_sub['total_export_costs'] = transport_sub['total_sea_freight_costs'] + transport_sub['total_road_transport_costs']

transport_sub.head()
transport_sub.to_csv('transport_sub.csv')
```
