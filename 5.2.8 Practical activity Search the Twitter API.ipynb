```python
# connect to the Twitter API with your OAuth authentication.

# Copy the YAML file and your Twitter keys over before you start to work.
import yaml
from yaml.loader import SafeLoader
from twitter import *

# Import the YAML file - remember to specify the whole path.
twitter_creds = yaml.safe_load(open('twitter_tmp.yaml', 'r').read())

# Pass your twitter credentials.
twitter_api = Twitter(auth=OAuth(twitter_creds['access_token'],
                                 twitter_creds['access_token_secret'], 
                                 twitter_creds['api_key'],
                                 twitter_creds['api_secret_key'] ))
```

```python
# 1. Connecting to Twitter and Confirm your connection.
print(twitter_api)

# Run a test with #python.
python_tweets = twitter_api.search.tweets(q='#python')

# View output.
print(python_tweets)
```

```python
# 2. Search trends worldwide for the IDs of the required cities (US as New York and UK as London).

# Identify worldwide trends.
trends_worldwide = twitter_api.trends.available()

# How many trends are available.
print(len(trends_worldwide))

# Example of trends_worldwide. # dict format
trends_worldwide[0]
```

```python
# Write a Twitter request to identify a city.
list_of_names = [_['name'] for _ in trends_worldwide]

# List of first 10 cities.
list_of_names[0:10]
```

```python
# a. Find New York and London.
city_1 = 'New York'

# Create a variable with "for if statement".
list_of_names_city_1 = [_ for _ in trends_worldwide if _['name'] == city_1]

# View the output.
print(len(list_of_names_city_1))

# Use index to find City.
list_of_names_city_1[0]

city_2 = 'London'

# Create a variable with "for if statement".
list_of_names_city_2 = [_ for _ in trends_worldwide if _['name'] == city_2]

# View the output.
print(len(list_of_names_city_2))

# Use index to find City.
list_of_names_city_2[0]
```

```python
# b. List of 'where on earth identifiers' (woeid).
list_of_names_city_1[0]['woeid']
list_of_names_city_2[0]['woeid']
```

```python
# c. Find a list of the trends specific to each city and save it as a Pandas DataFrame.

# Look at common trends in NY.
ny_trends = twitter_api.trends.place(_id = list_of_names_city_1[0]['woeid'])

# View the output.
ny_trends

# View output as a DataFrame. (MUST create a dataframe before saving as a csv.)
# Import Pandas.
import pandas as pd

# Create a DataFrame.
ny_trends_pd = pd.DataFrame(ny_trends[0]['trends'])

# View the DataFrame.
ny_trends_pd.head()
```

```python
# c. Find a list of the trends specific to each city and save it as a Pandas DataFrame.

# Look at common trends in London.
ldn_trends = twitter_api.trends.place(_id = list_of_names_city_2[0]['woeid'])

# View the output.
ldn_trends

# View output as a DataFrame. (MUST create a dataframe before saving as a csv.)
# Import Pandas.
import pandas as pd

# Create a DataFrame.
ldn_trends_pd = pd.DataFrame(ldn_trends[0]['trends'])

# View the DataFrame.
ldn_trends_pd.head()
```

```python
# Narrow te list down to 50,000 tweets NY
ny_trends_50k_pd = ny_trends_pd[ny_trends_pd['tweet_volume'] > 50000]\
.sort_values('tweet_volume', ascending=False)

# View the output.
print(ny_trends_50k_pd.shape)
print(ny_trends_50k_pd.head())

# Narrow te list down to 50,000 tweets London
ldn_trends_50k_pd = ldn_trends_pd[ldn_trends_pd['tweet_volume'] > 50000]\
.sort_values('tweet_volume', ascending=False)

# View the output.
print(ldn_trends_50k_pd.shape)
print(ldn_trends_50k_pd.head())
```

```python
# e. Save the output as a CSV file.
ny_trends_50k_pd.to_csv('ny_trends_50k_pd.csv', index = False)
ldn_trends_50k_pd.to_csv('ldn_trends_50k_pd.csv', index = False)
```

## Find the common trends between New York and London

```python
# 3.a.b.c
# get the trends for Cape Town in a JSON format with an indent of 4
import json

# Search for Cape Town.
ny_trends = twitter_api.trends.place(_id=2459115)

# View JSON output.
print(json.dumps(ny_trends, indent = 4))

# Search for Cape Town.
ldn_trends = twitter_api.trends.place(_id=44418)

# View JSON output.
print(json.dumps(ldn_trends, indent = 4))
```

```python
# Find common topics.
ny_trends_list = [trend['name'] for trend in ny_trends[0]['trends']]

# View the output.
print(ny_trends_list)

# Find common topics.
ldn_trends_list = [trend['name'] for trend in ldn_trends[0]['trends']]

# View the output.
print(ldn_trends_list)
```

```python
# Create a list of the common topics and use the set() function to determine the intersection.
ny_trends_list_set = set(ny_trends_list)
ldn_trends_list_set = set(ldn_trends_list)

com_set = ny_trends_list_set.intersection(ldn_trends_list_set)

com_set
```
